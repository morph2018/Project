# open函数
- open函数负责打开文件，带有很多参数
- 第一个参数： 必须有，文件的路径和名称
- mode：表明文件用什么方式打开
    - r:以只读方式打开
    - w：写方式打开，会覆盖以前的内容，如果没有文件，则创建
    - x：创建方式打开，如文件已经存在，报错
    - a：append方式，以追加的方式对文件内容进行写入
    - b： binary方式，二进制方式写入
    - t： 文本方式打开
    - +： 可读写



```python
# 打开文件，用写的方式
# r表示后面字符串内容不需要转义
# f称之为文件句柄
f = open(r"test01.txt", 'w')
# 文件打开后必须关闭
f.close()

# 此案例说明，以写方式打开文件，默认是如果没有文件，则创建
```

# with 语句
- with语句使用的技术是一种成为上下文管理协议的技术(ContextManagementProtocal)
- 自动判断文件的 作用域， 自动关闭不在使用的打开的文件句柄


```python
#with案例
with open(r'text01.txt','r') as f:
    pass
# 下面语句块开始对文件f进行操作
# 在本模块中不需要在使用close关闭文件f
```

```python
count = 1
while count:#只要条件变量的值不为0或者None或者''空字符串，都会继续循坏
    if count >= 5:
        print(count,'is not less than 5')
        break
    print(count,'is less than 5')
    count = count + 1 
```

```python
# with案例
with open(r'test01.txt', 'r') as f:
    # 按行读取内容
    strline = f.readline()
    # 此结构保证能够完整读取文件直到结束
    print(strline+'这是第一行')
    strline = f.readline()#执行f.readline()这句话用来跳入下一行，并返回下一行的值
    print(strline+'这是第二行')
    while strline:
        print(strline)
        strline = f.readline()
print(strline*100+'哈哈哈')#f.readline()会读完所有内容后返回一个''空字符串        
```

# seek（offset， from)
- 移动文件的读取位置，也叫读取指针
- from的取值范围：
    - 0： 从文件头开始偏移
    - 1：从文件当前位置开始偏移
    - 2： 从文件末尾开始偏移
- 移动的单位是字节(byte)
- 一个汉子由若干个字节构成
- 返回文件只针对当前位置



```python
# seek案例
# 打开文件后，从第5个字节出开始读取

# 打开读写指针在0处， 即文件的开头
with open(r'test01.txt', 'r') as f:
    print(f)#f现在是一个被'cp936'编解码器编了码的一大段byte码
    # seek移动单位是字节
    d = f.seek(5, 0)#从f最开头删除前面5个字节的byte码
    #l = list(f)
    #read函数会调用'gbk'编解码器对f中的中文字符的第二个字节进行解码，因此会报错。为避免报错，应让其对中文字符的两个字符一起解码。
    strChar = f.read()
    print(strChar)
```

```python
b ='月亮'
a = b.encode('unicode-escape')
print(a)
print('\u6708\u4eae')
c = b.encode('utf-8')
print(c)
print('\xe6\x9c\x88\xe4\xba\xae')
print('\u6708')
d = ord('月')
print(d)
```

```python
# 关于读取文件的练习
# 打开文件，三个字符一组读出内容，然后显示在屏幕上
# 每读一次，休息一秒钟

# 让程序暂停，可以使用time下的sleep函数
import time

with open(r'test01.txt', 'r') as f:
    # read参数的单位是字符，可以理解成一个汉字就是一个字符
    strChar = f.read(1)
    while strChar:
        print(strChar,end='')
        # sleep参数单位是秒
        time.sleep(0.2)
        strChar = f.read(1)
# 作业：
# 解释以下运行结果，为什么不是每行三个字符
```

```python
# tell函数： 用来显示文件读写指针的当前位置
with open(r'test01.txt', 'r') as f:
    strChar = f.read(3)
    pos = f.tell()
    
    while strChar:
        print(strChar)
        print(pos)
        
        strChar = f.read(3)
        pos = f.tell()
        
# 以下结果说明：
# tell的返回数字的单位是byte字节数
# read是以字符为单位
#换行符'\n'占2个字节，‘ ’空字符串占一个字节
```

# 文件的写操作-write
- write(str): 把字符串写入文件
- writeline(str): 把字符串按行写入文件
- 区别：
    - write函数参数只能是字符串
    - writerline参数可以是字符串，也可以是字符序列



```python
# write 案例
# 1. 向文件追加一句诗

# a代表追加方式打开(append)
with open(r'test02.txt', 'a') as f:
    # 注意字符串内含有换行符
    f.write("生活不仅有眼前的苟且,\n还有远方的苟且")

```

```python
# 可以直接写入行， 用writelines
# writelines表示写入很多行，参数可以是list格式

# a代表追加方式打开
with open(r'test02.txt', 'a') as f:
    # 注意字符串内含有换行符
    f.writelines("生活不仅有眼前的苟且")
    f.writelines("还有远方的枸杞")
```

```python
#writerline参数可以是字符串，也可以是字符序列
l = ["I", "love", "wangxiaojing"]

# a代表追加方式打开
with open(r'test02.txt', 'w') as f:
    # 注意字符串内含有换行符
    f.writelines(l)
```

# 持久化 - pickle
- 序列化（持久化，落地）：把程序运行中的信息保存在磁盘上
- 反序列化： 序列号的逆过程
- pickle： python提供的序列化模块
- pickle.dump:序列化
- pickle.load:反序列化



```python
# 序列化案例
import pickle

age = 19
#将age以二进制的形式写入text02当中，并保存在磁盘内
with open(r'test02.txt', 'wb') as f:
    b = pickle.dump(age, f)
    print(b)
```

```python
# 反序列化案例

import pickle
#从磁盘内以二进制的形式读出text02的信息
with open(r'test02.txt', 'rb') as f:
    age = pickle.load(f)
    print(age)
```

```python
# 序列化案例2
import pickle

a = [19, 'liudana', "i love wangxiaojing", [185, 76]]

with open(r'test02.txt', 'wb') as f:
    pickle.dump(a, f)
```

```python
with open(r'test02.txt', 'rb') as f:
    a  = pickle.load(f)
    print(a)
```

# 持久化-shelve
- 持久化工具
- 类似字典，用kv对保存数据，存取方式跟字典也类似
- shelve.open打开之后必须关闭,否则写入的数据会遗失



```python
# 使用shelve创建文件并使用
import shelve

# 打开文件
# shv相当于一个字典
shv = shelve.open(r'shv.db')

shv['one'] = 1
shv['two'] = 2
shv['three'] = 3

shv.close()

# 通过以上案例发现，shelve自动创建的不仅仅是一个shv.db文件，还包括其他格式文件
```

```python
# shelve读取案例
shv = shelve.open(r'shv.db')

try:
    print(shv['one'])
    print(shv['threee'])
#Exception是所有错误的父类    
except Exception as e:
    print(e)
    
finally:
    shv.close()

print(shv)
```

# shelve特性
- 不支持多个应用并行写入
- 为了读取多个程序，open的时候可以使用flag=r
- 写回问题
    - shelv恶魔人情况下不会等待持久化对象进行任何修改
    - 解决方法： 强制写回：writeback=True



```python
# shelve 以只读打开
import shelve

shv = shelve.open(r'shv.db', flag='r')

try:
    k1 = shv['one']
    print(k1)
finally:
    shv.close()
```

```python
import shelve

 
shv = shelve.open(r'shv.db')
try:
    shv['one'] = {"eins":1, "zwei":2, "drei":3}
finally:
    shv.close()


shv = shelve.open(r'shv.db')
try:
    one = shv['one']
    print(one)
finally:
    shv.close()
```

```python
# shelve忘记写回案例
shv = shelve.open(r'shv.db')
try:
    k1 = shv['one']
    print(k1)
    # 此时，一旦shelve关闭，则内容还是存在于内存中，没有写回数据库
    k1["eins"] =100
finally:
    shv.close()
    
    
shv = shelve.open(r'shv.db')
try:
    k1 = shv['one']
    print(k1)
finally:
    shv.close()
```

```python
# shelve忘记写回，需要使用强制写回writeback=True
shv = shelve.open(r'shv.db', writeback=True)
try:
    k1 = shv['one']
    print(k1)
    k1["eins"] =100
finally:
    shv.close()
    
    
shv = shelve.open(r'shv.db')
try:
    k1 = shv['one']
    print(k1)
finally:
    shv.close()
```

```python
# shelve 使用with管理上下文环境,这样就不用每次写入数据都要关闭。

with shelve.open(r'shv.db', writeback=True) as shv:
    k1 = shv['one']
    print(k1)
    k1["eins"] =1000

    
    
with shelve.open(r'shv.db') as shv:
    print(shv['one'])
```
